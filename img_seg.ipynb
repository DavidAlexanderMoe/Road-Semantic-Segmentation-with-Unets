```python
import shutil



def organize_images(source_folder, unmasked, masked):

    # Create target directories if they don't exist
    if not os.path.exists(unmasked):
        os.makedirs(unmasked)
    if not os.path.exists(masked):
        os.makedirs(masked)

    # Traverse through the source folder
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            # Get the full path of the file
            file_path = os.path.join(root, file)

            # Check if the file is a PNG or JPEG
            if file.lower().endswith('.png'):
                # Move PNG file to the target PNG folder
                shutil.move(file_path, os.path.join(masked, file))
                print(f"Moved {file} to {masked}")
            elif file.lower().endswith('.jpeg') or file.lower().endswith('.jpg'):
                # Move JPEG file to the target JPEG folder
                shutil.move(file_path, os.path.join(unmasked, file))
                print(f"Moved {file} to {unmasked}")
```

```python
# Define source and target folders
source_folder = 'train'
unmasked = 'original'
masked = 'masked'

# Organize the images
organize_images(source_folder, unmasked, masked)
```
